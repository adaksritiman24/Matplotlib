```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
#distribution Graphs
#1---->DistPlot
#2----->Joint Plot
#3----->Pair Plot
#4----->Rug Plot

#Categorical Plot
#-----> Barplot
#-----> Count plot
#----->Box plot
#----->violin plot

#Matrix Plot
#---->Heat  map
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill']) #more like histogram and also gives kernel density estimation
plt.plot()
```

```python
sns.jointplot(tips['tip'],tips['total_bill'], kind = 'kde')
```

```python
sns.jointplot(tips['tip'],tips['total_bill'], kind = 'hex')
```

```python
sns.jointplot(tips['tip'],tips['total_bill'], kind = 'reg')
```

```python
sns.pairplot(tips, hue = 'sex',diag_kind='kde')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
tips.head()
```

```python
sns.barplot(x ='day',y ='tip',data = tips , hue ='smoker')
```

```python
sns.countplot(tips['smoker'])
```

```python
sns.countplot(tips['smoker'],hue ='sex',data =tips)
```

```python
sns.boxplot('total_bill',data = tips, hue ='sex')
```

```python
sns.boxplot(x=tips['sex'],y =tips['tip'])
```

```python
sns.violinplot(x=tips['day'],y=tips['tip'],split=True)
```
